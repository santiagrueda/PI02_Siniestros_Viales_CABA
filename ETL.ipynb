# ETL (Extracción, Transformación y Carga de datos)

##### El objetivo del proyecto es situarse como un Data analyst, es importante tener en cuenta que este rol es aquel que NO modifica ni transforma datos, lo que hace es interpretar y convertir esa data otorgada en información, por lo que este proceso de ETL no se hará ninguna transformación, sólo unas modificaciones para que sea más limpio y fácil de leer la data

##### Importo librerías necesarias 

```python
import pandas as pd
import numpy as np
```

##### Poseemos dos archivos Excel 'homicidios' y 'lesiones' los cuales contiene varias pestañas con información, para este caso, los datos de los cuales vamos a realizar un análisis, estan en las pestañas 'HECHOS' y 'VICTIMAS' de ambos datasets, por lo que explicitamente trabajaremos con esta data.

```python
hom_hechos = pd.read_excel('datasets/homicidios.xlsx', sheet_name='HECHOS')
hom_victimas = pd.read_excel('datasets/homicidios.xlsx', sheet_name='VICTIMAS')
les_hechos = pd.read_excel('datasets/lesiones.xlsx',sheet_name='HECHOS')
les_victimas = pd.read_excel('datasets/lesiones.xlsx',sheet_name='VICTIMAS')
```

### Tratamiento de Dataframe hom_hechos

```python
# Chequeo incial

hom_hechos.head(10)
```

##### Con el fin de normalizar el dataframe, convierto los nombres de las columnas todos a minúsculas

```python
hom_hechos.columns = hom_hechos.columns.str.lower()
```

##### Para comprender mejor las columnas 'pos x' y 'pos y', cambiamos su nombre por longitud y latitud

```python
hom_hechos.rename(columns={'pos x':'longitud', 'pos y':'latitud'})
```

### Tratamiento de Dataframe hom_victimas

```python
# Chequeo inicial

hom_victimas.head()
```

##### Con el fin de normalizar el dataframe, convierto los nombres de las columnas todos a minúsculas

```python
hom_victimas.columns = hom_victimas.columns.str.lower()
```

##### Cambio el nombre de 'id_hecho' a 'id' con el fin de mantener nombres de columnas similares entre dataframes

```python
hom_victimas.rename(columns={'id_hecho':'id'},inplace=True)
```

```python
# Chequeo

hom_victimas.head()
```

### Tratamiento de Dataframe les_hechos

```python
# Chequeo inicial

les_hechos.head()
```

##### Gracias a las indicaciones encontradas en el diccionario de datos, se establece que los valores nulos de la columna 'gravedad' pueden ser establecidos como de gravedad leve. Para el resto de columnas no encontramos ninguna indicación de cambio

```python
les_hechos['gravedad'].fillna('leve',inplace=True)
```

##### Renombro algunas columnas para mantener similitud entre nombres de columnas 

```python
les_hechos.rename(columns={'latutid':'latitud','franja_hora':'hh'},inplace=True)
```

### Tratamiento de Dataframe les_victimas

```python
# Chequeo inicial

les_victimas.head()
```

##### Gracias a las indicaciones en el diccionario de datos, se establece que los valores nulos de la columna 'gravedad' pueden ser establecidos como de gravedad leve. Para el resto de columnas no encontramos ninguna indicación de cambio

```python
les_victimas['GRAVEDAD'].fillna('leve',inplace=True)
```

##### Renombro algunas columnas para mantener similitud entre nombres de columnas 

```python
# Modifico los nombres de las columnas a minúsculas

les_victimas.columns = les_victimas.columns.str.lower()

les_victimas.rename(columns={'id hecho':'id','aaa':'aaaa'})
```

### Concatenación de Dataframes


##### Ahora que hicimos el proceso de ETL para cada archivo a trabajar, es posible realizar una concatenación entre los archivos similares y así trabajar de manera mas ordenada y ágil ya que poseemos la información en una cantidad menor de archivos

```python
# Concatenación de archivos por la pestaña 'hechos'

hechos = pd.concat([hom_hechos, les_hechos], ignore_index=True)

# Concatenación de archivos por la pestaña 'victimas'

victimas = pd.concat([hom_victimas, les_victimas], ignore_index=True)
```

```python
# Guardo mis archivos concatenados en un csv

hechos.to_csv('Datasets/CSV/hechos.csv', index=False)

victimas.to_csv('Datasets/CSV/victimas.csv', index=False)
```
